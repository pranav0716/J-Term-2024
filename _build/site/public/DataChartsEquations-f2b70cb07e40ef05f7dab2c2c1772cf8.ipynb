```python
#| label: scatter_plot
# Assuming you have a DataFrame named 'plain_pizza_df'
# Replace 'plain_pizza_df' with your actual DataFrame

# Sample data (replace with your actual data)
data = {
    'Year': plain_pizza_nyp['Year'],
    'Price': plain_pizza_nyp['Price as number'],  # Replace with the correct column name for price
}

df = pd.DataFrame(data)

# Calculate the average price for each year
average_prices = df.groupby('Year')['Price'].mean().reset_index()

# Create a scatter plot using Plotly Express
fig = px.scatter(
    average_prices,
    x='Year',
    y='Price',
    labels={'Price': 'Average Price of Plain Pizza Slice'},
    title='Average Price of Plain Pizza Slice (2014-2023)',
)

# Show the plot
fig.show()
```
